## Normality test
From the plot above we found the data matches a normal distribution well. To statistically determine whether the data is drawn from a normal distribution, we calculate the p value. We compare the p value to a significance level of 0.001. If the p value is smaller than this level, we can say the data is drawn from a normal distribution.

```python
_, pval = stats.normaltest(count_data)
print('p value:',pval)
```

Thus we conclude the data is drawn from a normal distribution (P<0.001).
